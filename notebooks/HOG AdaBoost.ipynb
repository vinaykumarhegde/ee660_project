```python
import numpy as np
from sklearn.svm import SVC
from sklearn import grid_search
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
import datasets
```

```python
data=datasets.load_cifar_hog()
scaler = StandardScaler()
X_train = scaler.fit_transform(data.X_train)
Y_train = data.y_train
X_test = scaler.transform(data.X_test)
Y_test = data.y_test
```

```python
rndf=RandomForestClassifier(n_estimators=15,max_features="log2",n_jobs=-1,random_state=1,verbose=1,warm_start=True)
adaboost=AdaBoostClassifier(rndf,n_estimators=50,random_state=1)
```

```python
%time adaboost.fit(X_train,Y_train)
print 'Accuracy: ', adaboost.score(X_test,Y_test)
```

```python
rndf=RandomForestClassifier(n_estimators=25,max_features="sqrt",n_jobs=-1,random_state=1,warm_start=True)
adaboost=AdaBoostClassifier(rndf,n_estimators=50,random_state=1)
```

```python
%time adaboost.fit(X_train,Y_train)
print 'Accuracy: ', adaboost.score(X_test,Y_test)
```

```python
rndf=RandomForestClassifier(n_estimators=25,max_features="sqrt",n_jobs=-1,random_state=1,warm_start=True)
adaboost=AdaBoostClassifier(rndf,n_estimators=100,random_state=1)
%time adaboost.fit(X_train,Y_train)
print 'Accuracy: ', adaboost.score(X_test,Y_test)
```

```python
rndf=RandomForestClassifier(n_estimators=100,max_features="sqrt",n_jobs=-1,random_state=1,warm_start=True)
adaboost=AdaBoostClassifier(rndf,n_estimators=50,random_state=1)
%time adaboost.fit(X_train,Y_train)
print 'Accuracy: ', adaboost.score(X_test,Y_test)
```
